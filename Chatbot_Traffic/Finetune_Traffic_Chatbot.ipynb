```python
import torch

file_path = "/usr/local/lib/python3.11/dist-packages/transformers/pytorch_utils.py"

with open(file_path, "r") as f:
    lines = f.readlines()

# Tìm dòng bắt đầu của hàm
start_line = None
for i, line in enumerate(lines):
    if line.strip().startswith("def id_tensor_storage"):
        start_line = i
        break

if start_line is not None:
    # Nội dung mới của hàm
    new_function = [
        "def id_tensor_storage(tensor: torch.Tensor) -> tuple[torch.device, int, int]:\n",
        '    """\n',
        "    Unique identifier to a tensor storage. Multiple different tensors can share the same underlying storage.\n",
        "    This identifier is guaranteed to be unique and constant for this tensor's storage during its lifetime.\n",
        "    Two tensor storages with non-overlapping lifetimes may have the same id.\n",
        '    """\n',
        "    DTensor = type(None)\n",
        "\n",
        "    if is_torch_greater_or_equal_than_2_0:\n",
        "        try:\n",
        "            from torch.distributed.tensor import DTensor\n",
        "        except ImportError:\n",
        "            try:\n",
        "                from torch.distributed._tensor import DTensor\n",
        "            except ImportError:\n",
        "                pass\n",
        "\n",
        "        if isinstance(tensor, DTensor):\n",
        "            local_tensor = tensor.to_local()\n",
        "            return tensor.device, local_tensor.storage().data_ptr(), tensor.nbytes\n",
        "\n",
        "    if tensor.device.type == 'xla' and is_torch_xla_available():\n",
        "        import torch_xla\n",
        "        unique_id = torch_xla._XLAC._xla_get_tensor_id(tensor)\n",
        "    else:\n",
        "        unique_id = storage_ptr(tensor)\n",
        "\n",
        "    return tensor.device, unique_id, storage_size(tensor)\n"
    ]

    # Tìm dòng kết thúc của hàm cũ để thay thế
    end_line = start_line + 1
    while end_line < len(lines):
        if lines[end_line].strip().startswith("def ") and end_line != start_line:
            break
        end_line += 1

    # Thay thế nội dung cũ bằng nội dung mới
    lines = lines[:start_line] + new_function + lines[end_line:]

    # Ghi đè lại file
    with open(file_path, "w") as f:
        f.writelines(lines)

    print("✅ Đã thay thế hoàn toàn hàm id_tensor_storage.")
else:
    print("❌ Không tìm thấy hàm id_tensor_storage để sửa.")
```

```python
def ensure_string(value):
    if isinstance(value, list):
        # Nếu là danh sách dict, chuyển dict thành chuỗi
        return "\n".join(
            [json.dumps(v, ensure_ascii=False) if isinstance(v, dict) else str(v) for v in value]
        )
    elif isinstance(value, dict):
        return json.dumps(value, ensure_ascii=False)
    elif isinstance(value, str):
        return value
    return str(value)  # fallback

def convert_to_prompt_format(topic_data):
    messages = []

    for topic, items in topic_data.items():
        for item in items:
            title = item.get("title", "")
            description = ensure_string(item.get("description", ""))
            causes = ensure_string(item.get("causes", ""))
            mechanism = ensure_string(item.get("mechanism", ""))
            meaning = ensure_string(item.get("meaning", ""))

            # Message 1: Giới thiệu
            l_messages = []
            system_message = {
                "from": "system",
                "value": "Bạn là một trợ lí Tiếng Việt nhiệt tình và trung thực. Hãy luôn trả lời một cách hữu ích nhất có thể, đồng thời giữ an toàn. Nếu một câu hỏi không có ý nghĩa hoặc không hợp lý về mặt thông tin, hãy giải thích tại sao thay vì trả lời một điều gì đó không chính xác, vui lòng không chia sẻ thông tin sai lệch."
            }
            user_message = {
                "from": "human",
                "value": f"Giới thiệu về nội dung: {title}?"
            }
            assistant_message = {
                "from": "gpt",
                "value": f"Mô tả của {title} là {description}"
            }
            l_messages.extend([system_message, user_message, assistant_message])
            messages.append(l_messages)

            # Message 2: Nguyên nhân
            l_messages = []
            user_message = {
                "from": "human",
                "value": f"Nguyên nhân dẫn đến vấn đề trong {title} là gì?"
            }
            assistant_message = {
                "from": "gpt",
                "value": f"Nguyên nhân:\n{causes}"
            }
            l_messages.extend([system_message, user_message, assistant_message])
            messages.append(l_messages)

            # Message 3: Cơ chế / giải pháp
            l_messages = []
            user_message = {
                "from": "human",
                "value": f"Cơ chế hoặc giải pháp cho {title} là gì?"
            }
            assistant_message = {
                "from": "gpt",
                "value": f"Cơ chế hoặc giải pháp:\n{mechanism}"
            }
            l_messages.extend([system_message, user_message, assistant_message])
            messages.append(l_messages)

            # Message 4: Ý nghĩa
            l_messages = []
            user_message = {
                "from": "human",
                "value": f"Ý nghĩa hoặc tác động của {title} là gì?"
            }
            assistant_message = {
                "from": "gpt",
                "value": f"Ý nghĩa:\n{meaning}"
            }
            l_messages.extend([system_message, user_message, assistant_message])
            messages.append(l_messages)

    return messages
```

```python
def read_json_and_convert_to_txt(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile:
        disease_data = json.load(infile)

    messages = convert_to_prompt_format(disease_data)

    with open(output_file, 'w', encoding='utf-8') as outfile:
        outfile.write(json.dumps(messages, ensure_ascii=False, indent=2))
```
